```python
def book_seat(seat, booked, total):
    if seat < 1 or seat > total:
        return "Invalid seat number!"
    if seat in booked:
        return f"Seat {seat} is already booked."
    booked.append(seat)
    return f"Seat {seat} booked successfully."

def cancel_seat(seat, booked):
    if seat not in booked:
        return f"Seat {seat} is not booked."
    booked.remove(seat)
    return f"Seat {seat} cancelled successfully."

def available_seats(total, booked):
    return [s for s in range(1, total+1) if s not in booked]


total_seats = int(input("Enter total number of seats in cinema hall: "))
booked_seats = []

while True:
    print("\n--- Cinema Booking Menu ---")
    print("1. Book a seat")
    print("2. Cancel a seat")
    print("3. Show available seats")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        seat = int(input("Enter seat number to book: "))
        print(book_seat(seat, booked_seats, total_seats))

    elif choice == 2:
        seat = int(input("Enter seat number to cancel: "))
        print(cancel_seat(seat, booked_seats))

    elif choice == 3:
        print("Available seats:", available_seats(total_seats, booked_seats))

    elif choice == 4:
        break

    else:
        print("Invalid choice! Please try again.")
```
