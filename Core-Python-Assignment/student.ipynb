```python
class Student:
    def __init__(self, data):
        self.data = data
        self.avg_all = {}

    def average(self):

        for student, marks in self.data.items():
            total = sum(marks)
            avg = total / len(marks)
            self.avg_all[student] = round(avg, 2)
        return self.avg_all  

    def top_performer(self):
        if not self.avg_all:   
            self.average()
        top_student = max(self.avg_all, key=self.avg_all.get)
        return top_student, self.avg_all[top_student]  #

students ={}
n = int(input("Enter number of students: "))

for i in range(n):
    name = input(f"\nEnter name of student {i+1}: ")
    m = int(input(f"Enter number of subjects for {name}: "))
    marks = []
    for j in range(m):
        mark = float(input(f"Enter mark {j+1} for {name}: "))
        marks.append(mark)
    students[name] = marks

obj = Student(students)
averages = obj.average()
top_name, top_avg = obj.top_performer()

print("\nAverage Marks:", averages)
print(f"Top Performer: {top_name} with {top_avg}")
```
