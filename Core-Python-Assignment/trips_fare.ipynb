```python
def calculate_fare(distance):
    base_fare = 50
    distance_fare = 10 * distance
    return base_fare + distance_fare

trips = [5, 10, 3]  

total = 0
for d in trips:
    fare = calculate_fare(d)
    print(f"Trip with {d} km: Fare = ${fare}")
    total += fare

print(f"\nTotal Fare for all trips: ${total}")
```
