```python
def cart(items):
    if len(items) is 0:
        return "cart is Empty!"
    sum=0
    for item in items:
        sum += items[item]
    if len(items)>5:
        sum -= sum*10/100
    return sum
cart_items = {}
n = int(input("Enter number of items: "))

for i in range(n):
    item = input(f"Enter name of item {i+1}: ")
    price = float(input(f"Enter price of {item}: "))
    cart_items[item] = price

print(cart(cart_items))
```
