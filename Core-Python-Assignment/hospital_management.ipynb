```python
class Patient:
    def __init__(self):
        self.records = []

    def add_patient(self, name, age, disease):
        self.records.append({"Name": name, "Age": age, "Disease": disease})

    def search_by_disease(self, disease):
        result = []
        for p in self.records:   
            if p["Disease"].lower() == disease.lower():
                result.append(p["Name"])
        return result
hospital = Patient()

n = int(input("Enter number of patients: "))

for i in range(n):
    print(f"\nEnter details for patient {i+1}:")
    name = input("Name: ")
    age = int(input("Age: "))
    disease = input("Disease: ")
    hospital.add_patient(name, age, disease)
search_disease = input("\nEnter disease to search: ")
found = hospital.search_by_disease(search_disease)

if found:
    print(f"Patients with {search_disease}: {found}")
else:
    print(f"No patients found with {search_disease}")
```
